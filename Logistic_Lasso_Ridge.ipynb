```python
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
iris = datasets.load_iris()
# print(iris['DESCR'])
#features of iris dataset
print(iris['feature_names'],'\n')
X = iris['data']
y = iris['target']

print("Shape of X (feature data):", X.shape)
print("Shape of y (target):", y.shape)
print()

logit = LogisticRegression(max_iter = 10000)
print(logit.fit(X,y))
print(logit.score(X,y))

# Using test train split method
from sklearn import metrics #for checking the model accuracy
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=100)
log = LogisticRegression()
log.fit(X_train,y_train)
prediction = log.predict(X_test)
print('The accuracy of the Logistic Regression is',metrics.accuracy_score(prediction,y_test))

```

## Changing the regularization parameter C in logistic Regression

```python
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
iris = datasets.load_iris()
# print(iris['data'][2:6])
print(iris['data'].shape)
X = iris['data']
y = iris['target']
log = LogisticRegression(max_iter=1000)
c=[2.5,.5,.75,1.75,1.5,1.25]

scores=[]
for k in c:
    log.set_params(C=k)
    log.fit(X,y)
    scores.append(log.score(X,y))
print(scores)
    
```

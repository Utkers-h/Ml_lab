```python
import mglearn
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split

X,y = mglearn.datasets.load_extended_boston()
X.shape
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=1,test_size=.3)

ridge=Ridge()
ridge.fit(X_train,y_train)
# ridge.fit(X_test,y_test)
```

```python
print("test score ",ridge.score(X_train,y_train))
print("Training score: ", ridge.score(X_test,y_test))
```

```python
# Changing alpha in ridge regression
ridge_01 = Ridge(alpha=0.1)
ridge_01.fit(X_train,y_train)
print("test score ",ridge_01.score(X_train,y_train))
print("Training score: ", ridge_01.score(X_test,y_test))
```

```python
ridge_10 = Ridge(alpha=10)
ridge_10.fit(X_train,y_train)

print("test score ",ridge_10.score(X_train,y_train))
print("Training score: ", ridge_10.score(X_test,y_test))
```

```python
import matplotlib.pyplot as plt
plt.plot(ridge.coef_,'s',label='Ridge_alpha =1 ')
plt.plot(ridge_10.coef_,'^',label='Ridge_alpha =10 ')
plt.plot(ridge_01.coef_,'v',label='Ridge_alpha =.1 ')
plt.legend()
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=1)
lasso.fit(X_train,y_train)
lasso.fit(X_test,y_test)
```

```python
print("Accuracy score for training: ", lasso.score(X_train,y_train))
print("Accuracy score for testing: ", lasso.score(X_test,y_test))
import numpy as np
print('No of features used: ',np.sum(lasso.coef_!=0))
```

```python
lasso_001 = Lasso(alpha=.001,max_iter=1000)
lasso_001.fit(X_train,y_train)
lasso_001.fit(X_test,y_test)
```

```python
print("Accuracy score for training: ", lasso_001.score(X_train,y_train))
print("Accuracy score for testing: ", lasso_001.score(X_test,y_test))
print('No of features used: ',np.sum(lasso_001.coef_!=0))
```

```python
lasso_01 = Lasso(alpha=.01,max_iter=1000)
lasso_01.fit(X_train,y_train)
lasso_01.fit(X_test,y_test)
```

```python
print("Accuracy score for training: ", lasso_01.score(X_train,y_train))
print("Accuracy score for testing: ", lasso_01.score(X_test,y_test))
print('No of features used: ',np.sum(lasso_01.coef_!=0))
```

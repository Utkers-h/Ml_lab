# Bouston housing dataset


```python
import pandas as pd
import matplotlib.pyplot as plt
from mglearn.datasets import load_extended_boston
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import seaborn as sns  

```

```python
boston = load_extended_boston()
boston[0].shape
X= boston[0]
y=boston[1]
X.shape
y.shape
```

```python
#Now standardising the data
boston_st = StandardScaler()
x_scaled = boston_st.fit_transform(X)
```

```python
#Now reducing the dimensions to 4,3,2 componenets
pca_2 = PCA(n_components=2)
pca_3 = PCA(n_components=3)
pca_4 = PCA(n_components=4)

#Fitting the data
pca_2.fit(x_scaled)
pca_3.fit(x_scaled)
pca_4.fit(x_scaled)

# Transforming the data
transformed_2=pca_2.transform(x_scaled)
transformed_3=pca_3.transform(x_scaled)
transformed_4=pca_4.transform(x_scaled)

#Now printing the transformed data shape
print("2 components: ", transformed_2.shape)
print("3 components: ", transformed_3.shape)
print("4 components: ", transformed_4.shape)
```

```python
#Visualising the data for 2 dimensions
sns.scatterplot(x=transformed_2[:,0],y=transformed_2[:,1],hue=y,palette="viridis")
plt.xlabel("Principal Component 1 ")
plt.ylabel("Principal Component 2 ")
```

```python
#Printing explained variance ratio
print("Explained variance ratio for 4 dimensions: ", pca_4.explained_variance_ratio_)
```

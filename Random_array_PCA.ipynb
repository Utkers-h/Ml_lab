```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
#Now creating random number array
np.random.seed(1) # so that we generate the same random numbers array each time
data = np.random.rand(1000,50)
data.shape
```

```python
data[0][3]
```

```python
# standardize and fitting the data
scaled_data = StandardScaler().fit_transform(data)
```

```python
# Now reducing it to 2 components
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)
print("Reduced data set shape : " ,pca_data.shape)
```

```python
#creating a data frame
df = pd.DataFrame(data=pca_data,columns=['Principal Component 1','Principal Component 2'])
df.head(5)
```

```python
sns.scatterplot(data=df,x='Principal Component 1',y='Principal Component 2')
```
